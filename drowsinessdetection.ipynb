```python
import torch
import cv2
import numpy as np
from matplotlib import pyplot as plt
from pygame import mixer
import os
```

```python
# Yolov5'ten custom modeli yükle
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')
```

```python
model
```

```python
mixer.init()
alarm_sound = mixer.Sound("alarm.wav")
alarm_on = False
```

```python
cap = cv2.VideoCapture(0)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame)
    labels = results.pandas().xyxy[0]['name'].tolist()

    if 'drowsy' in labels:
        if not alarm_on:
            print("UYKULU DURUM ALGILANDI! 🔴")
            alarm_sound.play(-1)
            alarm_on = True
        cv2.putText(frame, 'DROWSY', (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,0,255), 2)
    else:
        if alarm_on:
            print("Normal duruma dönüldü. ✅")
            alarm_sound.stop()
            alarm_on = False
        cv2.putText(frame, 'AWAKE', (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0), 2)

    cv2.imshow('Detection', np.squeeze(results.render()))

    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
